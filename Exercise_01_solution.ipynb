```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-2*np.pi, 2*np.pi, 100)
ys = np.sin(xs)
zs = np.cos(xs)
```

```python
plt.figure(figsize=(8,3))
plt.plot(xs, ys, 'r')
plt.plot(xs, zs, 'b')
plt.legend(['Sin(x)', 'Cos(x)'])
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0), 
plt.title('Sine and cosine functions from -2π to 2π')
plt.yticks([-1,0,1])
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], ['-2π', '-π', '0', 'π', '2π'])
plt.axhline(0, c='k', alpha=0.2)
plt.savefig('trig.png')
```
